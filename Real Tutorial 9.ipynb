```python
import pandas as pd
```

```python
wichita = pd.read_csv("https://datajournalism.tech/wp-content/uploads/2019/10/wichita.csv")
```

```python
wichita.info()
```

```python
wichita.head()
```

```python
wichita.subject_race != "other/unknown"
```

```python
wichita[wichita.subject_race != "other/unknown"]
```

```python
four = wichita[wichita.subject_race != "other/unknown"]
```

```python
four.info()
```

```python
race = four.subject_race.value_counts().reset_index()
```

```python
race.info()
```

```python
race.head()
```

```python
race = race.rename(columns = {"index": "subject_race", "subject_race": "count"})
```

```python
race.info()
```

```python
race.head()
```

```python
population = pd.DataFrame({"subject_race": ["white", "black",  "asian/pacific islander", "hispanic",], "number_people": [246343, 42679, 19272, 63659]}, columns=["subject_race", "number_people"])
```

```python
stops = pd.merge (race, population, on = "subject_race")
```

```python
stops["stop_rate"] = stops["count"]/stops["number_people"]*100
```

```python
stops
```

```python
stops = stops.sort_values("stop_rate", ascending=False)
```

```python
%matplotlib inline
```

```python
chart = stops.stop_rate.plot.barh()
chart.set_yticklabels(stops.subject_race)
```

```python
stops.to_csv("stops.csv")
```
